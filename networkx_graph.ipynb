```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load your data (replace 'your_data.csv' with your actual file)
nodes_data = pd.read_csv('patdf.csv')
edges_data = pd.read_csv('edges.csv')

# Create a graph
G = nx.Graph()

# Add nodes to the graph
for index, row in nodes_data.iterrows():
    G.add_node(row['Patient_ID'], age=row['Age'], gender=row['Gender'], diagnosis=row['Prev_Diag'])

# Add edges to the graph
for index, row in edges_data.iterrows():
    G.add_edge(row['Source'], row['Target'], type=row['Type'])

# Visualization (optional)
options = {
    'node_color': 'lightblue',
    'node_size': 50,
    'linewidths': 0,
    'width': 0.1,
    'with_labels': False,
    'font_size': 8,
}

pos = nx.spring_layout(G, iterations=15, seed=1721)
nx.draw(G, pos, **options)
plt.show()

```
